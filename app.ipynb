In summary, when dealing with a symmetric correlation matrix, you can choose to analyze either the upper or lower triangle to avoid redundancy and simplify the interpretation of the relationships between variables. This simplification is commonly done to focus on unique correlation pairs and is particularly useful when dealing with large datasets. The choice between the upper or lower triangle doesn't impact the information captured; the key is to be consistent in your approach.

```python
# now lets do some tasks. do some operations so that we get onlu upper half of the values 


np.triu(np.ones_like(defaulter.corr()))
```

```python

defaulter_corr.where(np.triu(np.ones_like(defaulter.corr()).astype(bool),k=1))

# removing NaN values 

def_corr_unst=defaulter_corr.where(np.triu(np.ones_like(defaulter.corr()).astype(bool),k=1)).unstack()

def_corr_unst=def_corr_unst.reset_index().rename(columns={'level_0':'var','level_1':'var2',0:'corr'})

def_corr_unst=def_corr_unst.dropna(subset='corr').sort_values(by='corr',ascending=False)


# The difference between the two snippets is that the first one drops any row with at least one NaN value in any column,
# while the second one specifically drops rows where the 'corr' column has NaN values. If you want to drop rows based on 
#NaN values in the 'corr' column, you should use the second snippet with the corrected syntax.



def_corr_unst.head()
```
